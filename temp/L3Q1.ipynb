# Question 1

Write a program to extract the contents (excluding any tags) from the following five
websites
- https://en.wikipedia.org/wiki/Web_mining
- https://en.wikipedia.org/wiki/Data_mining
- https://en.wikipedia.org/wiki/Artificial_intelligence
- https://en.wikipedia.org/wiki/Machine_learning
- https://en.wikipedia.org/wiki/Mining
save the content in five separate .doc file. Considering a vector space model and do
the following operations according to the query “Mining of large data”
- Bag-of-Words (Document set)
- TF (Document set)
- IDF (Document set)
- TF-IDF (Document set)
- TF-IDF (Query)
- Normalized (Query)
- Normalized - TF-IDF (Document set)
- Cosine Similarity
- Euclidean Distance

## TF - IDF

```python
## TF of Documents
len1 = len(filtered_sentence1)
len2 = len(filtered_sentence2)
len3 = len(filtered_sentence3)
len4 = len(filtered_sentence4)
len5 = len(filtered_sentence5)

tf = [[l[i], filtered_sentence1.count(l[i])/len1, filtered_sentence2.count(l[i])/len2, filtered_sentence3.count(l[i])/len3, filtered_sentence4.count(l[i])/len4, filtered_sentence5.count(l[i])/len5] for i in range(len(l))]
print("TF of Documents\n")
print(tabulate(tf,headers=["Term","Doc1","Doc2","Doc3","Doc4","Doc5"]))
```

```python
## IDF of Documents
idf = []

for i in range(len(l)):
    if(result[i][1]>0 and result[i][2]>0):
        idf.append(np.log(3/2))
    else:
        idf.append(np.log(3))
        
# print(idf)
```

```python
## TF-IDF of Document Set
tfidf = tf
for i in range(len(l)):
    tfidf[i][1] = tf[i][1] * idf[i]
    tfidf[i][2] = tf[i][2] * idf[i]
    tfidf[i][3] = tf[i][3] * idf[i]
    tfidf[i][4] = tf[i][4] * idf[i]
    tfidf[i][5] = tf[i][5] * idf[i]

print("\n\nTF-IDF of Documents\n")
print(tabulate(tfidf,headers=["Term","Doc1","Doc2","Doc3","Doc4","Doc5"]))
```

## Distance and Ranking

```python
cs1 = 0
cs2 = 0
cs3 = 0
cs4 = 0
cs5 = 0
ed1 = 0
ed2 = 0
ed3 = 0
ed4 = 0
ed5 = 0

for i in range(len(l)):
    cs1+=(tfidf[i][1]*tfidf2[i][1])
    cs2+=(tfidf[i][2]*tfidf2[i][1])
    cs3+=(tfidf[i][3]*tfidf2[i][1])
    cs4+=(tfidf[i][4]*tfidf2[i][1])
    cs5+=(tfidf[i][5]*tfidf2[i][1])

    ed1= ed1 + (tfidf[i][1]-tfidf2[i][1])**2
    ed2= ed2 + (tfidf[i][2]-tfidf2[i][1])**2
    ed3= ed3 + (tfidf[i][3]-tfidf2[i][1])**2
    ed4= ed4 + (tfidf[i][4]-tfidf2[i][1])**2
    ed5= ed5 + (tfidf[i][5]-tfidf2[i][1])**2
    
    
ed1 = math.sqrt(ed1)
ed2 = math.sqrt(ed2)
ed3 = math.sqrt(ed3)
ed4 = math.sqrt(ed4)
ed5 = math.sqrt(ed5)

cosine = {"doc1": cs1, "doc2": cs2, "doc3": cs3, "doc4": cs4,"doc5": cs5}

euclidean_distance = {"doc1": ed1, "doc2": ed2, "doc3": ed3, "doc4": ed4, "doc5": ed5}
```
