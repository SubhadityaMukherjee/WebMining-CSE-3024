```python
import pandas as pd
import sklearn as sk
import nltk
import re
import os
from tabulate import tabulate
from nltk.stem import WordNetLemmatizer 
  
lemmatizer = WordNetLemmatizer() 

stopwords = nltk.corpus.stopwords.words('english')
```

# Hierarchical Clustering (Use Euclidian Distance Function)
- Health-News-Tweets.zip

```python
!ls 'data/Dataset/Health-News-Tweets/Health-Tweets/'
```

```python
docs = os.listdir('data/Dataset/Health-News-Tweets/Health-Tweets/'); docs
```

```python
total_docs = len(docs); total_docs
```

```python
docs_open = {}
for a in docs:
    temp = open(f'data/Dataset/Health-News-Tweets/Health-Tweets/{a}','r',errors = 'ignore').readlines()
    temp = [re.sub('http:[a-zA-Z0-9\/.\-\']*','',x.split('|')[-1]).strip() for x in temp]
    temp = [re.sub('[^A-Za-z0-9]+', ' ', x) for x in temp]
    docs_open[a.split('.')[0]] = temp
```

```python
docs_open['bbchealth']
```

```python
for a in docs_open:
    docs_open[a] = [x for x in docs_open[a] if x not in stopwords]
    docs_open[a] = [lemmatizer.lemmatize(x).lower() for x in docs_open[a]]
    
```

```python
docs_open['bbchealth']
```

```python
corpus = []
for a in docs_open:
    temp = docs_open[a]
    [corpus.extend(x.split(' ')) for x in temp]
```

```python
len(corpus)
```

```python
corpus_un = list(set(corpus)); len(corpus_un)
```

```python
corpus_undic = dict(zip(corpus_un,[list('0'*16) for i in range(len(corpus_un))])); corpus_undic
```

```python
for a in list(corpus_undic.keys()):
    if not str(a).isalpha():
        del corpus_undic[a]
```

```python
len(corpus_undic)
```

```python
pd.DataFrame.from_dict(corpus_undic)
```

#  Decision Tree

- a) F - Score
- b) ROC & AUC Curve
- c) R2
- d) Error Rate
- e) Accuracy



```python
health = pd.read_csv('data/Dataset/diabetes/diabetes_simplied.csv')
```

```python
health
```

```python
for a in health.columns:
    print(a, health[a].isna().sum())
```

```python
health['Outcome'].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(health.drop(['Outcome'],axis=1), health['Outcome'], test_size=0.33, random_state=42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
